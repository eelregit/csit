```python
Omega_m = 0.3156
Omega_L = 0.6844
lambdaz = 0.01
factor = 1.

def Hubble(a):
    matter = Omega_m/a**3
    Lambda = Omega_L
    return np.sqrt(matter+Lambda)

def growth_int(a):
    return 1./(a*Hubble(a))**3

def growth_D(a):
    factor = 5*Omega_m*Hubble(a)/2
    return factor*integrate.quad(growth_int, 0, a)[0]

def rel_D(z):
    a = 1/(1+z)
    return growth_D(a)/growth_D(1)
```
